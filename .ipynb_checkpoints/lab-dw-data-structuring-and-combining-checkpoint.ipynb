```python
import pandas as pd

url = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv"
insurance_company_df1 = pd.read_csv(url)
insurance_company_df1
```

```python
import pandas as pd

url = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv"
insurance_company_df2 = pd.read_csv(url)
insurance_company_df2
```

```python
import pandas as pd

url = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv"
insurance_company_df3 = pd.read_csv(url)
insurance_company_df3
```

```python
insurance_company_combined_df = pd.concat([insurance_company_df1, insurance_company_df2, insurance_company_df3],axis=0)
insurance_company_combined_df.reset_index(drop=True, inplace=True)
insurance_company_combined_df
```

```python
insurance_company_combined_df.drop_duplicates(subset=['Customer'], inplace=True)
insurance_company_combined_df.reset_index(drop=True, inplace=True)
print(insurance_company_combined_df)
```

```python
#Cleaning Functions: The clean_data function handles various cleaning tasks, including renaming columns, handling percentage signs, standardizing gender values, replacing state abbreviations, and extracting middle values.
#Printing the Result: Finally, the cleaned and combined DataFrame is printed.

print(insurance_company_combined_df.dtypes)
```

```python
#including renaming columns
insurance_company_combined_df.rename(columns={"Customer": "Customer_ID"}, inplace=True)
print(insurance_company_combined_df)
```

```python
#handling percentage signs

insurance_company_combined_df['Customer Lifetime Value'] = insurance_company_combined_df['Customer Lifetime Value'].str.replace('%', '')
```

```python
print(insurance_company_combined_df["Customer Lifetime Value"])
```

```python
#standardizing gender values
gender_mapping = {'F': 'F', 'Female': 'F', 'female': 'F', 'Femal': 'F','M': 'M', 'Male': 'M', 'male': 'M'}
insurance_company_combined_df.Gender = insurance_company_combined_df['Gender'].replace(gender_mapping)
insurance_company_combined_df.Gender.value_counts()
```

```python
#replacing state abbreviations
state_mapping = {"AZ": "Arizona", "Cali": "California", "WA": "Washington"}
insurance_company_combined_df.ST = insurance_company_combined_df["ST"].replace(state_mapping)
insurance_company_combined_df.ST.value_counts()
```

```python
#extracting middle values

print("Unique values in 'Number of open complaints':") 
print(insurance_company_combined_df["Number of Open Complaints"].unique())
```

```python
## Function to extract the middle value
def extract_middle_value(complaint):
    if isinstance(complaint, str):
         parts = complaint.split('/')
         if len(parts) == 3:
             return parts[1] 
         return complaint
```

```python
insurance_company_combined_df['Number of Open Complaints'] = insurance_company_combined_df['Number of Open Complaints'].apply(lambda x:x.split("/")[1] if type(x)== str else x)
```

```python
insurance_company_combined_df['Number of Open Complaints'] = insurance_company_combined_df['Number of Open Complaints'].apply(extract_middle_value)
```

```python
insurance_company_combined_df["Number of Open Complaints"] = pd.to_numeric(insurance_company_combined_df["Number of Open Complaints"])
```

```python
print(insurance_company_combined_df['Number of Open Complaints'])
```

2. Create a pivot table that shows the average customer lifetime value per gender and education level. Analyze the resulting table to draw insights.

```python

insurance_company_df4 = pd.DataFrame(insurance_company_df4)

# Create a pivot table
insurance_company_pivot = insurance_company_df4.pivot_table(
    values='total_claim_amount',
    index='sales_channel',
    aggfunc='sum'
)

# Round the results to 2 decimal points
insurance_company_pivot = insurance_company_pivot.round(2)

# Print the resulting pivot table
print(insurance_company_pivot)
```
